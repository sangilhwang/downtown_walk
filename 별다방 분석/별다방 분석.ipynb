```python
address_list = [i.text for i in soup.select("#mCSB_3_container > ul > li > p")]
```

```python
star_df = pd.DataFrame(address_list)
```

```python
star_df["address_list"] = star_df.iloc[:, 0].str.split()
```

```python
star_df["자치구"] = star_df["address_list"].map(lambda x: x[1])
```

```python
star_df2 = star_df.groupby("자치구").count()
```

```python
star_df2 = star_df2.reset_index()
```

```python
star_df2 = star_df2.iloc[:, :2]
```

```python
star_df2 = star_df2.rename(columns = {0 : "매장수"})
```

```python
star_df3 = pd.merge(seoul_district, star_df2, on = "자치구", how = "inner")
```

```python
star_df3 = star_df3.iloc[:, 1:]
```

```python
star_df3["매장수/면적"] = star_df3["매장수"] / star_df3["면적 (km²)"]
```
