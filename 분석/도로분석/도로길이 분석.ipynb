```python
import pandas as pd
```

```python
seoul_district = pd.read_csv("C:/Users/ITSC/Desktop/MiniPrj2/Git/Data/seoul_district.csv")
```

```python
road_df = pd.read_csv("C:/Users/ITSC/Desktop/MiniPrj2/Git/Data/도로+현황_20240317153112.csv", skiprows = 4)
```

```python
road_df = road_df.iloc[1:, 1:4]
```

```python
road_df
```

```python
road_df2 = road_df.sort_values(by = "자치구별(2)").loc[road_df["자치구별(2)"].isin(seoul_district["자치구"])].reset_index(drop = True)
```

```python
road_df2 = road_df2.rename(columns = {"자치구별(2)" : "자치구"})
```

```python
road_df2 = road_df2.iloc[:, :2]
```

```python
road_df2
```

```python
df_combined = pd.merge(seoul_district, road_df2, on = "자치구", how = "inner")
```

```python
df_combined = df_combined.iloc[:, 1:]
```

```python
df_combined["연장/면적"] = round(df_combined["연장"] / df_combined["면적 (km²)"], 2)
```
